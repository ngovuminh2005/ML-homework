```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
import random
import math

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

random_state = 0
```

```python
data = pd.read_csv('australian.dat', delim_whitespace=True, header=None)
data.head(15)
```

## 2. Decision Tree

### 2.1. Khảo sát các giá trị khác nhau của max_depth

```python

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

# Khảo sát các giá trị khác nhau của max_depth
max_depths = [i for i in range(1, 11)]
for depth in max_depths:
    # Tạo mô hình Decision Tree với max_depth hiện tại
    dtc = DecisionTreeClassifier(max_depth=depth, random_state=random_state)

    # Huấn luyện mô hình
    dtc.fit(X_train, y_train)

    # Dự đoán trên tập kiểm tra
    y_pred = dtc.predict(X_test)

    # Tính độ chính xác
    accuracy = accuracy_score(y_test, y_pred)

    print(f"Max depth: {depth}, Accuracy: {accuracy}")
```

### 2.2. Tìm kiếm siêu tham số

```python

from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'max_depth': [i for i in range(1, 11)],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

# Create a Decision Tree Classifier
dtc = DecisionTreeClassifier(random_state=random_state)

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=dtc, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

## 3. Random Forest

### 3.1 Khảo sát các giá trị khác nhau của n_estimators

```python

from sklearn.ensemble import RandomForestClassifier

# Khảo sát các giá trị khác nhau của n_estimators
n_estimators_values = [i for i in range(10, 101, 10)]
for n_estimators in n_estimators_values:
    # Tạo mô hình Random Forest với n_estimators hiện tại
    rfc = RandomForestClassifier(n_estimators=n_estimators, random_state=random_state)

    # Huấn luyện mô hình
    rfc.fit(X_train, y_train)

    # Dự đoán trên tập kiểm tra
    y_pred = rfc.predict(X_test)

    # Tính độ chính xác
    accuracy = accuracy_score(y_test, y_pred)

    print(f"N_estimators: {n_estimators}, Accuracy: {accuracy}")
```

### 3.2 Tìm kiếm siêu tham số

```python
# prompt: tìm siêu tham số từ random forest kể trên

# Define the parameter grid
param_grid = {
    'n_estimators': [i for i in range(10, 101, 10)],
    'max_depth': [i for i in range(1, 11)],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

# Create a Random Forest Classifier
rfc = RandomForestClassifier(random_state=random_state)

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```
